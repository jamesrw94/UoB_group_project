```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
twitter_1 = pd.read_csv("twitter_info2021-03-31 12:26:23.081651.csv")
twitter_2 = pd.read_csv("twitter_info2021-04-04 15:05:21.028058.csv")
twitter_3 = pd.read_csv("twitter_info2021-04-14 18:07:51.095378.csv")
```

```python
full_twitter = pd.concat([twitter_1,twitter_2,twitter_3])
full_twitter = full_twitter.drop_duplicates(subset='text')
```

```python
plot = full_twitter.groupby(["paper","term"]).count().sort_values(by='text')['text'].plot(kind="barh",figsize=(10,10))
plot.set_xlabel("number of tweets");
```

```python
fig, ax = plt.subplots(3, 2, figsize=(26, 20))
fig.subplots_adjust(hspace=.5, wspace=0.5)
ax = ax.ravel()
for ind, paper in enumerate(full_twitter['paper'].unique()):
    plot = full_twitter[full_twitter.paper==paper].groupby(["term"]).count().sort_values(by='text')['text'].plot(kind="barh",ax=ax[ind])

    ax[ind].set_title('distribution of topics in {}'.format(paper))
    ax[ind].set_xlabel("number of tweets")
    ax[ind].set_ylabel('topics')
```

```python
full_twitter.to_json("twitter_info.json",orient = "records", date_format = "epoch")
```

```python
papers=pd.DataFrame(full_twitter['paper'].unique())
```

```python
papers['correct_answers']=0
papers['incorrect_answers']=0
papers.columns = ['paper','correct_answers','incorrect_answers']
```

```python
papers.to_json("paper_answer_info.json",orient = "records", date_format = "epoch")
```
